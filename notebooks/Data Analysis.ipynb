```python
import pandas as pd, matplotlib.pyplot as plt, numpy as np, os, datetime as dt
```

```python
df = pd.read_csv("../data/interim/results_Season_2021.txt", names=["Day", "Mo_day", "Year", "Time", "Opponent", "Result", "Score_team", "Score_opp", "Ws", "Ls","Streak", "Home","OT", "teamcode", "gamecode", "Location_spc"])
```

```python
df.head()
```

```python
df.info()
```

```python
# post ALL-Star Break
df_post_all_star = df[df["Year"] > 2020]
```

```python
df_post_all_star
```

```python
df_post_all_star.Year = df_post_all_star.Year.astype(str)
```

```python
df_post_all_star["DT"] = df_post_all_star["Mo_day"] + " " + df_post_all_star["Year"]
```

```python
df_post_all_star.DT = pd.to_datetime(df_post_all_star.DT, format=" %b %d %Y")
```

```python
df_final = df_post_all_star[df_post_all_star.DT > dt.datetime(2021, 3, 4)]
```

```python
df_final = df_final.iloc[:, :-1]
```

```python
df_final = df_final.fillna(value="null")
```

```python
df_final.to_csv("../data/interim/results_Season_2021_postallstar.txt", index=False, header=False)
```
